### 单只股日内分时

```python
import akshare as ak

# 获取当前交易日的分时数据
stock_intraday_em_df = ak.stock_intraday_em(symbol="000571")
print(stock_intraday_em_df)

# 或使用新版接口
# stock_zh_a_minute_df = ak.stock_zh_a_minute(symbol="sz000571", period="1", adjust="qfq")
# print(stock_zh_a_minute_df)
```

```python
# 获取历史某日的分时数据
stock_zh_a_hist_min_em_df = ak.stock_zh_a_hist_min_em(
    symbol="000571",
    period="1",  # 1分钟数据
    start_date="2025-12-10 09:15:00",
    end_date="2025-12-10 15:00:00",
    adjust=""
)
print(stock_zh_a_hist_min_em_df)
```

```python
stock_zh_a_hist_df = ak.stock_zh_a_hist(
    symbol="000571",
    period="daily",
    start_date="20251201",
    end_date="20251210",
    adjust="hfq"
)
print(stock_zh_a_hist_df)
```
